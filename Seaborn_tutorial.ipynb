Seaborn is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive. https://www.geeksforgeeks.org/python-seaborn-tutorial/

```python
import seaborn as sns
import matplotlib.pyplot as plt

data=sns.load_dataset('iris')
sns.lineplot(x="sepal_length",y="sepal_width", data=data)
plt.title("Title using Matplotlib Function")
sns.set_style(style='ticks')
'''
There are five themes available in Seaborn.
darkgrid
whitegrid
dark
white
ticks'''
sns.despine()
plt.show()
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=data)
```

```python
sns.pairplot(data=data, hue='species')
plt.show()
```

```python
sns.barplot(x='species', y='sepal_width',data=data)
```

```python
sns.barplot(x='species',y='sepal_length',data=data)
```

```python
sns.boxplot(x='species',y='sepal_length',data=data)
```

```python
sns.violinplot(x='species',y='sepal_length',data=data)
```

```python
sns.stripplot(x='species',y='sepal_length',data=data)
```

```python
sns.swarmplot(x='species',y='sepal_length',data=data)
```

```python
sns.factorplot(x='species',y='sepal_length',data=data)
```

```python
sns.histplot(x='species',y='sepal_length',data=data)
```

```python
sns.distplot(data["sepal_width"])
```
