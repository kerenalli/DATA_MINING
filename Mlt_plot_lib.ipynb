```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,10])
y=np.array([0,10])
#Default
plt.plot(x,y)
#Without line
#plt.plot(x,y,'o')
#Multi Point
x1=np.array([0,1,2,3,4,5])
y1=np.array([0,1,4,9,16,25])
plt.plot(x1,y1)
plt.show()
```

```python
x1=np.array([0,1,2,3,4,5])
y1=np.array([0,1,4,9,16,25])
y2=np.array([0,1,8,27,64,125])
y3=np.array([0,1,3,6,10,15])
y4=np.array([0,1,2,3,4,5,6])
plt.plot(y1,'p:r')
plt.plot(y2, 's-b')
plt.plot(y3,'o--c')
plt.plot(y4, 'x-.g')
plt.show()
```

Matplotlib Labels


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,1,2,3,4,5])
y=np.array([0,1,4,9,16,25])
#Font Style
font1 = {'family':'serif','color':'red','size':15}
font2 = {'family':'serif','color':'blue','size':12}

#Plot Title
plt.title("Man vs Women", fontdict=font1, loc='center')
#X Axis Label
plt.xlabel("Man", fontdict=font2)
#Y-Label
plt.ylabel("women", fontdict=font2)

plt.plot(x,y,'o:b')
plt.show()
```

Multiple sub-plots

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,1,2,3])
y=np.array([0,1,4,9])
plt.title("Multi-plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(1,2,1)
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([0,1,8,27])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(1,2,2)
plt.plot(x,y)


plt.show()
```
